```python
import pandas as pd
import joblib
```

```python
df = pd.read_csv("../resources/shot_logs_clean.csv")
df.head()
```

```python
# Set features. This will also be used as your x values.
selected_features = df.drop(["outcome", "game_clock"], axis=1)
X = selected_features
y = df["outcome"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_train.head()
```

```python
# For each model, load file and use .score function to find accuracy score of model
files = [
        "logistic_regression_model.sav",
        "decision_tree_model.sav",
        "knn_model.sav",
        "random_forest_model.sav",
        "svm_model.sav"
        ]

rows = []
for file in files:
    loaded_model = joblib.load(file)
    result = loaded_model.score(X_test, y_test)
    rows.append([file, result])
```

```python
# Create DataFrame with model and relative score
model_df = pd.DataFrame(rows, columns=["Model", "Accuracy Score"]).sort_values(["Accuracy Score"],ascending=False)
model_df["Model"] = model_df["Model"].str.split(".").str[0]
model_df["Model"] = model_df["Model"].str.replace("_"," ")
model_df
```
