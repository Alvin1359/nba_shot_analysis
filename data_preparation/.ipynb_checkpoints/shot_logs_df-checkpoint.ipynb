```python
import pandas as pd
```

```python
# Read data from resources folder
shot_logs_df = pd.read_csv("../resources/shot_logs.csv")
shot_logs_df.head()
```

```python
# Convert all column names to lowercase
shot_logs_df.columns = shot_logs_df.columns.str.lower()
```

```python
# Select only relevant columns
shot_logs_df = shot_logs_df[["fgm", "period", "game_clock", "shot_clock", "dribbles", "touch_time", "shot_dist", "pts_type", "close_def_dist"]]
```

```python
# Rename necessary columns
shot_logs_df = shot_logs_df.rename(columns={"fgm":"outcome"})
```

```python
# Remove duplicates and rows with null values
shot_logs_df = shot_logs_df.drop_duplicates()
shot_logs_df = shot_logs_df.dropna()
shot_logs_df.head()
```

```python
# Export clean DataFrame to CSV file
shot_logs_df.to_csv("../resources/shot_logs_clean.csv", index=False)
```
