```python
import pandas as pd
```

```python
shot_18 = pd.read_csv("../resources/nba_shotchartdetail_2018-19_clean.csv")
shot_18.head()
```

```python
shot_08 = pd.read_csv("../resources/nba_shotchartdetail_2008-09_clean.csv")
shot_08.head()
```

```python
comparison_players = shot_18.merge(shot_08, how="inner", on="player_name")
comparison_players.head()
```

```python
comparison_players = comparison_players["player_name"].value_counts().rename_axis("player_name").reset_index(name="counts")
comparison_players = comparison_players["player_name"]
```

```python
comparison_players.to_csv("../resources/comparison_players.csv", index=False)
```
